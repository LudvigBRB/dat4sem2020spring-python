```python
#Ekstra
#henter diæten for dyr på bestemt side

import bs4
import requests

#url = 'http://www.prehistoric-wildlife.com/species/t/tiktaalik.html'
url = 'http://www.prehistoric-wildlife.com/species/s/samburupithecus.html'

html = requests.get(url)
txt = html.text
soup = bs4.BeautifulSoup(txt, 'html.parser')

#soup.find_all('p') # alle paragrafer tages
events = soup.find_all('p')[0]#.find_all('br') #kun første paragraf tages

something = events.text.split(':')[6].split('.')[0] # split teksten
print(something)
```
